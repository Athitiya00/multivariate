### Eaxple on Iris dataset

```python
import pandas as pd
```

```python
example_df = pd.read_csv('https://raw.github.com/pandas-dev/pandas/master/pandas/tests/io/data/csv/iris.csv')
example_df
```

### ลอง plot PCA

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components=2)
```

```python
new_pca = pca.fit_transform(example_df.iloc[:,:-1])
```

```python
from matplotlib import pyplot as plt
```

```python
plt.plot(new_pca[:50,0],new_pca[:50,1],'or',label='Setosa')
plt.plot(new_pca[50:100,0],new_pca[50:100,1],'og',label='Versicolor')
plt.plot(new_pca[100:,0],new_pca[100:,1],'oy',label='Virginica')
plt.legend()
```

###Distance Matrix (แสดงระยะห่างแต่ละจุดของเมทริค)

```python
#ใส่รูป
```

```python
from sklearn.metrics import pairwise_distances 
dis_matrix = pairwise_distances(example_df.iloc[:,:-1]) #Distance Matrix คือการเอาแต่ละจุดมาเทียบกัน
dis_matrix 
```

```python
plt.imshow(dis_matrix) #โชว์รูป ค่ามากสว่าง ค่าน้อยเข้ม 
# สีเข้ม คือ มีระยะห่างมีค่าน้อยเพราะเป็นสายพันธ์เดียวกัน(กลุ่มเดียวกันใกล้กันเอง) สีสว่าง คือ มีระยะห่างมีค่ามาก
# แถวแนวนอน1-50แรก สรุปได้ว่า พันธ์ที่1ห่างจากพันธุ์ที่2อยู่แต่พันธุ์ที่3ห่างกว่า
```

## ทำ MDS

```python
from sklearn.manifold import MDS
```

```python
mds = MDS(n_components=2)
```

```python
new_mds = mds.fit_transform(example_df.iloc[:,:-1])
new_mds
```

```python
plt.plot(new_mds[:50,0],new_mds[:50,1],'or',label='Setosa')
plt.plot(new_mds[50:100,0],new_mds[50:100,1],'og',label='Versicolor')
plt.plot(new_mds[100:,0],new_mds[100:,1],'oy',label='Virginica')
plt.legend() #มองดูว่าจุดไหนห่างจากจุดไหน 
#สิ่งที่เกิดขึ้นคือมันคำนวณระยะห่างระหว่างจุด ระยะห่างทุกคนเท่าเดิมแต่รูปไม่เหมือนกันเพราะเป็นการเลือกจุดไม่เหมือนกัน
```

```python
plt.plot(new_mds[:50,0],new_mds[:50,1],'or',label='Setosa')
plt.plot(new_mds[50:100,0],new_mds[50:100,1],'og',label='Versicolor')
plt.plot(new_mds[100:,0],new_mds[100:,1],'oy',label='Virginica')
plt.plot(new_mds[6,0],new_mds[6,1],'*k')
plt.plot(new_mds[56,0],new_mds[56,1],'*m')
plt.plot(new_mds[116,0],new_mds[116,1],'*b')
plt.legend() #เลือกจุดเพิ่มเพื่อดูความห่างระยะห่างจะเท่ากัน ทุกคนจะต้องเหมือนกัน
#ม่วงใกล้กับนำ้เงิน ดำห่างเพื่อน
```

### การใช้ Pre-computed Distance Matrix 

```python
dis_matrix = pairwise_distances(example_df.iloc[:,:-1],metric='cityblock') #cityblock
```

```python
mds2 = MDS(n_components=2,dissimilarity = 'precomputed')
```

```python
new_mds2 = mds2.fit_transform(dis_matrix)
```

```python
plt.plot(new_mds[:50,0],new_mds[:50,1],'or',label='Setosa')
plt.plot(new_mds[50:100,0],new_mds[50:100,1],'og',label='Versicolor')
plt.plot(new_mds[100:,0],new_mds[100:,1],'oy',label='Virginica')
plt.legend()
```

```python
plt.plot(new_mds2[:50,0],new_mds[:50,1],'or',label='Setosa')
plt.plot(new_mds2[50:100,0],new_mds[50:100,1],'og',label='Versicolor')
plt.plot(new_mds2[100:,0],new_mds[100:,1],'oy',label='Virginica')
plt.plot(new_mds2[6,0],new_mds[6,1],'*k')
plt.plot(new_mds2[56,0],new_mds[56,1],'*m')
plt.plot(new_mds2[116,0],new_mds[116,1],'*b')
plt.legend() 
```
